```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 500, 20)
y = -2 * x * x + 4 * x + 5

max_index = np.argmax(y)
min_index = np.argmin(y)
max_point = (x[max_index], y[max_index])
min_point = (x[min_index], y[min_index])

fig, ax = plt.subplots()

ax.scatter(max_point[0], max_point[1], color='red')
ax.scatter(min_point[0], min_point[1], color='red')

ax.annotate(f'Max: {max_point}', xy=max_point, xytext=(max_point[0], max_point[1] - 10), color='red')
ax.annotate(f'Min: {min_point}', xy=min_point, xytext=(min_point[0], min_point[1] + 10), color='red')

print(y)

ax.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
plt.grid(True)
plt.show()
```
